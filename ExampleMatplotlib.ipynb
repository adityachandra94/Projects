# Matplotlib

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("seaborn")    ##Set a style for figures
```

```python
x = np.linspace(-5, 5, 100)
y = x**2
z = y**2
```

```python
plt.plot(x,y,'b')
plt.xlabel("x")
plt.ylabel("$f(x)$")
```

```python
plt.plot(x,z,'--g')
plt.xlabel("x")
plt.ylabel("$f(x)$")
```

```python
plt.plot(x,y,'b')
plt.plot(x,z,'r')
plt.xlabel("x")
plt.ylabel("$f(x)$")
plt.legend(['$x^2$', '$y^2$'], loc='best')    #Plotting 2 graphs together
```

### Subplot for dividing figure into multiple plots

```python
plt.subplot(2,1,1)       #Subplot 2,1 creates two vertical divisions and 2,1,1 refers to 1st subplot
plt.plot(x,y)
plt.title('x vs y')

plt.subplot(2,1,2)       #Subplot 2,1,2 refers to 2nd subplot
plt.plot(z,x,'r')
plt.title('z vs x')        
```

```python
plt.subplot(2,2,1)            #Subplot 2,2 creates two side by side divisions and 2,2,1 refers to 1st subplot
plt.plot(x,y)
plt.title('x vs y')

plt.subplot(2,2,2)            #Subplot 2,2,2 refers to 2nd subplot
plt.plot(z,x,'r')
plt.title('z vs x') 
```

### Scatterplot

```python
x = np.random.rand(60)        
y = np.random.rand(60)
plt.scatter(x, y, s=85)       #s represents size of points
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Scatter plot of x vs y')
```

### Histogram

```python
nu = []
for i in range(50):
    nu.append(np.random.randint(1,50))
```

```python
plt.hist(nu, bins = 40)
plt.title('Histogram of Nu')
plt.xlabel('Nu')
plt.ylabel('Frequency')
```

### Barplot

```python
animals = ['Lion', 'Tiger', 'Giraffe', 'Elephant']
population = [853, 326, 786, 659]

plt.bar(animals, population, width = 0.5)
plt.xlabel('Animals')
plt.ylabel('Population')
plt.title('Population of animals')
```

### Pie Chart

```python
animals = ['Lion', 'Tiger', 'Giraffe', 'Elephant']
population = [853, 326, 786, 659]
focus = (0,0.1,0,0)                 #Explode population of tiger from the rest of the chart

plt.pie(population, labels = animals, explode = focus, startangle = 150)
plt.legend(title = 'Animals', loc = 'best')
plt.axis('equal')                   #Equal axis ensures pie chart is circular
plt.title('Animal Population')
```

### Box Plot

```python
rdf = pd.DataFrame(np.random.randn(3,3), columns=['A', 'B', 'C'])
plt.boxplot(rdf, labels=['A', 'B', 'C'])
```
